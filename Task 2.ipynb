```python
def batScaler(x):
    x['Batting_Strike_Rate'] = (x['Batting_Strike_Rate'] - x['Batting_Strike_Rate'].mean())/x['Batting_Strike_Rate'].std(axis=0)
    x['Batting_Average'] = (x['Batting_Average'] - x['Batting_Average'].mean())/x['Batting_Average'].std(axis=0)
    x['Runs_Scored'] = (x['Runs_Scored'] - x['Runs_Scored'].mean())/x['Runs_Scored'].std(axis=0)
    return x
    
    
```

```python
def bowlScaler(x):
    x['Economy'] = (x['Economy']-x['Economy'].mean())/x['Economy'].std(axis=0)
    x['Bowling_Average'] = (x['Bowling_Average']-x['Bowling_Average'].mean())/x['Bowling_Average'].std(axis=0)
    x['Bowling_Strike_Rate'] = (x['Bowling_Strike_Rate']-x['Bowling_Strike_Rate'].mean())/x['Bowling_Strike_Rate'].std(axis=0)
    return x 
```

```python
def allrounderScaler():
    x['Batting_Strike_Rate'] = (x['Batting_Strike_Rate'] - x['Batting_Strike_Rate'].mean())/x['Batting_Strike_Rate'].std(axis=0)
    x['Batting_Average'] = (x['Batting_Average'] - x['Batting_Average'].mean())/x['Batting_Average'].std(axis=0)
    x['Runs_Scored'] = (x['Runs_Scored'] - x['Runs_Scored'].mean())/x['Runs_Scored'].std(axis=0)
    x['Economy'] = (x['Economy']-x['Economy'].mean())/x['Economy'].std(axis=0)
    x['Bowling_Average'] = (x['Bowling_Average']-x['Bowling_Average'].mean())/x['Bowling_Average'].std(axis=0)
    x['Bowling_Strike_Rate'] = (x['Bowling_Strike_Rate']-x['Bowling_Strike_Rate'].mean())/x['Bowling_Strike_Rate'].std(axis=0)
    
```
