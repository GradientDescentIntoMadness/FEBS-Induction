We can observe that there are 141 values from the year 2022. And according to required statement, we are supposed to filter out data from this specific year alone. So let us filter using the year.

```python
cricket[cricket['Year']=='2022']
```

```python
cricket_new = cricket[cricket['Year']=='2022']
```

```python
cricket_new.to_csv("cricket_newdata.csv", index = False)
```
